```python
from moviepy.editor import *
```

```python
# Load the Drive helper and mount
from google.colab import drive
drive.mount('/content/drive')
```

```python
clip1 = VideoFileClip(r'/content/drive/MyDrive/moviepy/TrialTutorial.mp4')
clip2 = VideoFileClip(r'/content/drive/MyDrive/moviepy/speaker1.mp4')
clip2 = clip2.resize(0.25)
w1,h1 = clip1.size
w2,h2 = clip2.size
print(w1,h1)
print(w2,h2)
```

```python
trial_clip = CompositeVideoClip([clip1,
                                 clip2.set_position(("right","bottom"))
                                 .set_start(4).crossfadein(1)])
final_clip = trial_clip.subclip(t_end=(0,0,30))
```

```python
final_clip.write_videofile(r'/content/drive/MyDrive/moviepy/FinalVideo.mp4')
```
